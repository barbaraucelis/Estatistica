# Regressão Linear Simples

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
# Regressão - criação de conjunto x e y
x,y = make_regression(n_samples=200, n_features=1, noise=30)
```

```python
# Verificar o tamanho do vetor
len(y)
```

```python
# Gerar gráfico do plano cartesiano
plt.scatter(x,y)
```

```python
# Iniciar modelo de regressão
modelo = LinearRegression()
```

```python
# Rodar uma regressão linear com os valores de x e y
modelo.fit(x,y)
```

```python
# criando variáveis para os coeficientes linear e angular
coeficiente_angular = modelo.coef_
coeficiente_linear = modelo.intercept_
```

```python
print('Coeficiente Angular', coeficiente_angular)
print('Coeficiente Linear', coeficiente_linear)
```

```python
# Plotar gráfico dos pontos com a reta alinhada
plt.scatter(x,y)
plt.plot(x, coeficiente_linear + coeficiente_angular*x, color='red')
plt.scatter(2.5, coeficiente_linear + coeficiente_angular*2.5, color='purple', s=100)

plt.show()
```
